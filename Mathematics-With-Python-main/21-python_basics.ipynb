# 2 Python Basics
## 2.1 Python as a calculator
In a code cell program statements may be executed.
Similar to a calculator, the result of the calculation is displayed immediately.

```python
3+2
```

If the result of a calculation is assigned to a variable, no output appears.

```python
x = 3+2
```

The value of a variable can be output using the [print()](https://docs.python.org/3/library/functions.html#print) function.

```python
print(x)
```

Variables can be reused in other code cells.

```python
x+4
```

## 2.2 Arithmetic operators
Python has seven basic arithmetic operators.

$$
\begin{array}{|l|l|l|l|l|l|l|l|}
\hline
\mathsf{operator} & \mathsf{addition} & \mathsf{subtraction} & \mathsf{multiplication} & \mathsf{division} & \mathsf{floor} \, \mathsf{division}  & \mathsf{modulus}  & \mathsf{power} \\
\hline
\mathsf{priority} & 4 & 4 & 3 & 3 & 3 & 3 & 1 \\ 
\hline
\mathsf{symbol} & +  & -  & *  & /  & //  & \%  & ** \\
\hline
\end{array}
$$

```python
13 / 5
```

```python
13 // 5
```

```python
13 % 5
```

The precedence of the arithmetic operators follows standard mathematical conventions.

```python
3 + 2*4**2 
```


Each arithmetic operator can be combined with the assignment operator to form augmented assignment operators, also called compound assignment operators: $+=, -=, *=, /=, //=, \%=,$ and $**=$. These operators perform the arithmetic operation and assignment in one step.

```python
x = 3
x *= 2
print('x =', x)
```

Augmented assignment operators require a variable on the left side to store the result.
You cannot use them with constants or expressions.

```python
2*=2
```
